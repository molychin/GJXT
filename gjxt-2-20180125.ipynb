```python
"""
--ver 0.3/简化代码
--（应届）在籍性别人数/性别比例/各年级(2014级以后)
select xbm,njdm_id,count(*) from JW_XJGL_XSJBXXB where xjztdm='01' and njdm_id>='2014' group by xbm,njdm_id order by njdm_id,xbm
   	XBM	NJDM_ID	COUNT(*)
"""
import matplotlib.pyplot as plt

init_data=((1,1,2014,2204),
           (2,2,2014,1936),
           (3,1,2015,2079),
           (4,2,2015,2158),
           (5,1,2016,2456),
           (6,2,2016,2772),
           (7,1,2017,2740),
           (8,2,2017,2824))

num=4
mf_rate,grade=[],[]
order=range(num)

labels='male','female'
colors='yellowgreen','gold'
explode=0,0.1
fig = plt.figure(figsize=(10, 6))

for i in order:
    sum=init_data[i*2][3]+init_data[i*2+1][3]   #2014~2017
    grade.append(init_data[i*2][2])
    mf_rate.append([init_data[i*2][3]/sum,init_data[i*2+1][3]/sum])

    ax= fig.add_subplot(1,4,i+1)
    plt.axis('equal')
    ax.set_title(grade[i])
    ax.pie(mf_rate[i],explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=30)  

plt.show()
```
