```python
#测试显示学生成绩
#学生成绩图表显示
#coding=utf-8
from matplotlib.font_manager import FontProperties 
import matplotlib.pyplot as plt
print ('测试显示学生成绩-学生成绩图表显示')

font = FontProperties(fname=r"C:\\WINDOWS\\Fonts\\simsun.ttc", size=14)  #C:\WINDOWS\Fonts  

try:
    file_name='data/学生成绩-测试.csv'   #可以使用中文文件名 
    file_data=open(file_name,'r',encoding='utf-8')  #encoding='utf-8'必须加上，否则处理中文（utf-8编码）会出错
    row_data=file_data.readline()  #读出表头，以免影响后面的实际数据处理
    print (row_data)
    
    xx,yy=[],[]
    for row_data in file_data :
        record=row_data.split(',')
        #print (row_data,record)
        index=record[0]
        xn=record[1]
        xq=record[2]
        kch=record[3]
        xh=record[5]
        cj=record[7]
        #print (index,xn,xq,kch,xh,cj)    
        xx.append(float(index))
        yy.append(float(cj))
        
    #print(xx,'\n',yy)
    plt.grid()
    plt.plot(xx,yy)
    #plt.title('单个学生成绩分析')
    plt.title("单个学生成绩分析",fontproperties=font)
    plt.xlabel('课程门次',fontproperties=font)
    plt.ylabel('课程成绩',fontproperties=font)
    plt.show()
except IOError as err:
    print('File error.'+str(err))
finally:
    file_data.close()
```

```python
#测试显示学生成绩
#处理远程数据库数据（Oracle）
from __future__ import print_function
import cx_Oracle
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties 

print ('测试显示学生成绩-处理远程数据库数据（Oracle）')

font = FontProperties(fname=r"C:\\WINDOWS\\Fonts\\simsun.ttc", size=14)  #C:\WINDOWS\Fonts  
# Connect as user "hr" with password "welcome" to the "oraclepdb" service running on this computer.
connection = cx_Oracle.connect("gjfx","caca070706","10.28.200.188/orcl")
cursor = connection.cursor()
SQL="select xvh,bfzcj from zftal_jwlg.gjfx_view_cjb where jxb_id='16-2017-2)-A0930058-200500027-2'"
cursor.execute(SQL)

cj_set,cj_set1=[],[]
for record_data in cursor:
    cj=record_data[1]
    if (cj>=60):
        cj_set.append(cj)
    else:
        cj_set1.append(cj)

xx=range(len(cj_set))
xx1=range(len(cj_set1))
print (xx)

plt.grid()
#plt.plot(xh_set,cj_set)
plt.scatter(xx,cj_set,color='b',marker='^')
plt.scatter(xx1,cj_set1,color='r',marker='x')

plt.title("单门课程学生成绩分析",fontproperties=font)
plt.xlabel('序号',fontproperties=font)
plt.ylabel('课程成绩',fontproperties=font)

plt.show()
```
