```python
"""
雷达图显示单个学生的成绩 v1.0
"""
import molychin.database as mldb
import numpy as np
import matplotlib.pyplot as plt

print ('雷达图显示单个学生的成绩（Oracle）')
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签

cursor=mldb.moly_condb()    #连接数据库，使用默认用户名
xh='020611209'           #'050214214'
#SQL="select kcmc,bfzcj,zgcj,xnmmc,xqmmc,kch,kcxzmc from zftal_jwlg.gjfx_xscjb where xh='"+xh+"' order by xnmmc,xqmmc"
SQL="select kcmc,bfzcj,zgcj,xnmmc,xqmmc,kch,kcxzmc from zftal_jwlg.gjfx_xscjb where xh='"+xh+"' order by kcmc"
cursor.execute(SQL)
result_set=cursor.fetchall()

kcmc_list,cj_list,zgcj_list=[],[],[]
for record in result_set:
    kcmc=record[0]
    cj=record[1]
    zgcj=record[2]
    
    kcmc_list.append(kcmc)    #课程名称
    cj_list.append(cj)        #课程总评成绩
    zgcj_list.append(zgcj)    #重修、补考后的最高成绩

labels=np.array(kcmc_list)
data_cj=np.array(cj_list)
data_zgcj=np.array(zgcj_list)
dataLenth=len(labels)
angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
angles = np.concatenate((angles, [angles[0]])) # 闭合
cj_list.append(cj_list[0])
zgcj_list.append(zgcj_list[0])
#data = np.concatenate((data, [data[0]])) # 闭合
#print (cj_list)
#print (zgcj_list)

fig = plt.figure(figsize=[14,12])
ax = fig.add_subplot(111, polar=True,facecolor='lightcyan')# polar参数！！

#学年学期背景色显示
N = 8
theta = np.linspace(np.pi/N, 2 * np.pi+np.pi/N, N, endpoint=False)
radii = [93]*N
width=[2*np.pi/N]*N
#colors=np.random.rand(N)
colors=np.linspace(0,1,N)
bars = ax.bar(theta, radii, width=width, bottom=5.0)

for bar,col in zip(bars,colors):
    bar.set_facecolor(plt.cm.viridis(col))
    bar.set_alpha(0.9)

#增加及格线/优秀线显示
ang_jige=np.linspace(0,2*np.pi,100)
data_jige=[60]*100
data_youxiu=[90]*100
ax.plot(ang_jige,data_jige,c='r',linewidth=1)
ax.plot(ang_jige,data_youxiu,c='pink',linewidth=3)

#显示总评成绩点
ax.plot(angles,cj_list, 'bo-', linewidth=2)    #总评成绩
ax.fill(angles,cj_list, facecolor='coral', alpha=0.8)    # 填充
#显示最高成绩点
#ax.plot(angles,zgcj_list, 'bo-', linewidth=2)    #最高成绩
#ax.fill(angles,zgcj_list, facecolor='chocolate', alpha=0.8)    # 填充

ax.set_thetagrids(angles * 180/np.pi, labels)
ax.set_title("学生个人成绩", va='bottom')
ax.set_rlim(0,100)
ax.grid(True)

plt.show()
cursor.close()
```

```python
"""
雷达图显示单个学生的成绩 v1.2
v1.1 增加第一次考试成绩和最高成绩的对比
v1.2 课程学年学期背景色与实际课程相符
v1.2 按学年学期排列

"""
import molychin.database as mldb
import numpy as np
import matplotlib.pyplot as plt
import collections as clt

print ('雷达图显示单个学生的成绩（Oracle）')
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签

cursor=mldb.moly_condb()    #连接数据库，使用默认用户名
xh='080513104'           
#'050214214'/'020611209'/050213214/092216233/Z09414132/Z05215227/020315135/Z09414124/092213208/170115118/030115215
#172415129/072210102/061613134/080513104
SQL="select kcmc,min(bfzcj) mincj,min(xnmmc||'-'||xqmmc) minxnxq,greatest(max(bfzcj),max(cxcj)) maxcj, \
     max(xnmmc||'-'||xqmmc) maxxnxq from zftal_jwlg.gjfx_xscjb where xh='"+xh+"' group by kcmc order by minxnxq"
#print (SQL)
cursor.execute(SQL)
result_set=cursor.fetchall()

kcmc_list,min_cj_list,max_cj_list,min_xnxq_list,max_xnxq_list=[],[],[],[],[]
for record in result_set:
    kcmc=record[0]
    min_cj=record[1]
    min_xnxq=record[2]
    max_cj=record[3]
    max_xnxq=record[4]
    
    kcmc_list.append(kcmc)    #课程名称
    min_cj_list.append(min_cj)        #课程总评成绩
    min_xnxq_list.append(min_xnxq)
    max_cj_list.append(max_cj)    #重修、补考后的最高成绩
    max_xnxq_list.append(max_xnxq)

labels=np.array(kcmc_list)
dataLenth=len(labels)
angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
angles = np.concatenate((angles, [angles[0]])) # 闭合
min_cj_list.append(min_cj_list[0])
max_cj_list.append(max_cj_list[0])

fig = plt.figure(figsize=[14,12])
ax = fig.add_subplot(111, polar=True,facecolor='lightcyan')# polar参数！！

#学年学期背景色显示
#-----------------------------#
xnxq_cnt=clt.Counter(min_xnxq_list)
theta_list,xnxq_list,width_list=[],[],[]
theta_temp=0
count=0
for xnxq,value in xnxq_cnt.items():
    xnxq_list.append(xnxq)
    if count==0:
        theta_temp=value/2
    else:   
        theta_temp=theta_temp+value/2+width_list[count-1]/2
    theta_list.append(theta_temp)
    width_list.append(value)
    count=count+1    
    
N=len(xnxq_list)
#print (xnxq_list,theta_list,width_list)
theta_temp=2*np.pi/sum(width_list)
#print(sum(width_list))
#print (theta_temp)

radii = [95]*N
thate=[theta_temp*a for a in theta_list]
width=[theta_temp*a for a in width_list]
bars = ax.bar(thate, radii, width=width, bottom=5.0)
#给扇形区上色
colors=np.linspace(0,1,N)
for bar,col in zip(bars,colors):
    bar.set_facecolor(plt.cm.viridis(col))
    bar.set_alpha(0.9)
#------------------------------#
#增加及格线/优秀线显示
ang_jige=np.linspace(0,2*np.pi,100)
data_jige=[60]*100
data_youxiu=[90]*100
ax.plot(ang_jige,data_jige,c='r',linewidth=1)
ax.plot(ang_jige,data_youxiu,c='pink',linewidth=3)


#显示最高成绩点
ax.plot(angles,max_cj_list, 'b+-', linewidth=2)    #最高成绩
ax.fill(angles,max_cj_list, facecolor='#A9A9A9', alpha=0.7)    # 填充
#显示总评成绩点
ax.plot(angles,min_cj_list, 'yo-', linewidth=2)    #总评成绩
ax.fill(angles,min_cj_list, facecolor='coral', alpha=0.9)    # 填充

ax.set_thetagrids(angles * 180/np.pi, labels)
ax.set_title("学生个人成绩", va='bottom')
ax.set_rlim(0,110)
ax.grid(True)

plt.show()
cursor.close()
```
