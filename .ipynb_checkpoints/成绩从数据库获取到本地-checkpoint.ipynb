### 根据条件数据，逐步导出并追加数据到外部xscjb01.csv文件

```python
#耗时45分钟
cursor = mldb.moly_condb()
cond_df=pd.read_csv('data/cj_cond.csv')
counter=0    #如果导出数据中断，从中断的地方开始，使用counter=1；将cj_cond.csv中已完成数据导出的条件删除。

print(pd.datetime.now())
for index,row in cond_df.iterrows():
    records=[]
    index=row[0]
    xnm=str(row['XNM'])
    xqm=str(row['XQM'])
    xhh=str(row['XHH'])
    cnt=row['CNT']
    print(index,xnm,xqm,xhh,cnt)

    SQL1="select * from zftal_jwlg.gjfx_view_full_cjb where xnm='"+xnm+"' and xqm='"+xqm+"' and substr(xh_id,1,2)='"+xhh+"'"
    cursor.execute(SQL1)
    print(cursor.statement)    #打印执行的语句

    for reco in cursor:
        record_list=[]
        for ii in range(len(reco)):
            record_list.append(reco[ii])
        records.append(record_list)

    if (counter==0):
        field_name_list=[]
        for field in cursor.description:
            field_name_list.append(field[0])
        xscjb_df=pd.DataFrame(records,columns=field_name_list)       
        xscjb_df.to_csv('data/xscjb02.csv',index=False)   
        counter=1
    else:
        xscjb_df=pd.DataFrame(records)
        xscjb_df.to_csv('data/xscjb02.csv',mode='a',header=0,index=False)   #追击数据模式，头部字段不用追加
    
    print('ok')

print(pd.datetime.now())
```
