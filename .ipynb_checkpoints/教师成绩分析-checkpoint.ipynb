## 教师成绩分析
★★★★★单个教师课程成绩分布分析
1、对教师的各科成绩进行绘图；
2、对成绩形态进行分类，并作出说明；
3、对各类成绩形态给出提高教学质量的改进意见和建议；


```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

#%matplotlib inline
plt.rcdefaults()
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['figure.figsize']=[9,6] #改变显示图形的大小
```

```python
#从本地数据文件中读取成绩数据
file_name='data/xscjb20190215.csv'
fields=['XNM','XQM','JSGH','JSXM','KCH','KCMC','BFZCJ']
df_cjb=pd.read_csv(file_name,usecols=fields,dtype={'XNM':np.object,'XQM':np.object})
df_cjb.head()
```

```python
df_cjb.info()   #查看字段类型/3810319
```

```python
jsxm='高燕'   #教师姓名：沈健/秦志刚/赵根林/高燕/宋东兴/陈希棠/邹永林/周思林/张根华/徐惠钢/殷旭东/钱斌
df_cjb_single=df_cjb[(df_cjb.JSXM==jsxm) & (df_cjb.XNM=='2018') & (df_cjb.XQM=='3')]
df_cjb_single.head()
```

```python
df_cjb_single.to_csv('data/gaoyan.csv',index=False)
```

### 单个教师课程成绩分析

#### 数据预处理
1、统计每门课程的成绩人数

```python
df_cjb_group=df_cjb.groupby(['JSXM','KCMC','BFZCJ']).count()
df_cjb_group.reset_index(inplace=True)
df_cjb_group.head()
```

```python
jsxm='秦志刚'   #教师姓名：沈健/秦志刚/赵根林/高燕/宋东兴/陈希棠/邹永林/周思林/张根华/徐惠钢/殷旭东/钱斌
df_cjb_single=df_cjb_group[df_cjb_group.JSXM==jsxm]
df_cjb_single
```

#### 分析某个教师的课程成绩分布

```python


#绘制图像    
point_size=80     #图像点的大小
len1=len(df_cjb_single.groupby('KCMC').count())

df_cjb_single.insert(0,'SCALE',df_cjb_single.KCH*point_size)  #按比例显示人数
#df_jscjb_single.eval('KCH=KCH*23',inplace=True)
plt.figure(figsize=[2*len1,10])
plt.grid()

plt.scatter('KCMC','BFZCJ','SCALE',data=df_cjb_single)
plt.plot([-1,len1],[60,60],'r-.')
plt.xticks(rotation=30)
plt.yticks(range(0,101,5))

plt.title("单个教师全部课程成绩分析")
plt.xlabel('课程名称')
plt.ylabel('课程成绩')

pic_file_name='img/'+jsxm+'任课成绩分析.png'
#plt.savefig(pic_file_name)
plt.show()
```
